```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a sample dataset
np.random.seed(0)
data_size = 10
variable1 = np.random.rand(data_size) * 100  # Values between 0 and 100
variable2 = np.random.rand(data_size) * 100  # Values between 0 and 100, treated as percentages

# Create buckets for Variable 1
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df = pd.DataFrame({
    'Variable 1': pd.cut(variable1, bins=bins, labels=labels),  # Binning Variable 1
    'Variable 2 (%)': variable2
})

# Create pivot table for heatmap
pivot_table = df.pivot_table(index='Variable 1', values='Variable 2 (%)', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap: Variable 1 (Buckets) vs Variable 2 (%)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a sample dataset
np.random.seed(0)
data_size = 100  # Increase the data size for better granularity
variable1 = np.random.rand(data_size) * 100  # Values between 0 and 100
variable2 = np.random.rand(data_size) * 100  # Values between 0 and 100

# Create more granular buckets for Variable 1 and Variable 2
bins_var1 = np.linspace(0, 100, 11)  # 10 bins for Variable 1
bins_var2 = np.linspace(0, 100, 11)  # 10 bins for Variable 2

labels_var1 = [f'{int(bins_var1[i])}-{int(bins_var1[i+1]-1)}' for i in range(len(bins_var1)-1)]
labels_var2 = [f'{int(bins_var2[i])}-{int(bins_var2[i+1]-1)}' for i in range(len(bins_var2)-1)]

# Create DataFrame with binned values
df = pd.DataFrame({
    'Variable 1': pd.cut(variable1, bins=bins_var1, labels=labels_var1),  # Binning Variable 1
    'Variable 2': pd.cut(variable2, bins=bins_var2, labels=labels_var2)   # Binning Variable 2
})

# Create a pivot table with counts of occurrences for each combination of Variable 1 and Variable 2 bins
pivot_table = df.pivot_table(index='Variable 1', columns='Variable 2', aggfunc='size', fill_value=0)

# Convert counts to percentages of total values
pivot_table_percentage = pivot_table / pivot_table.sum().sum() * 100

# Ensure all rows and columns are aligned by filling missing columns with 0
pivot_table_percentage = pivot_table_percentage.fillna(0)

# Custom annotation formatting with percentage symbol, handling missing values or invalid data
def annotate_percentage(data):
    # Create a matrix of formatted string annotations with percentage sign, handling missing/invalid values
    formatted_data = []
    for row in data.values:  # Use .values to ensure we're working with matrix-like data
        formatted_row = []
        for val in row:
            formatted_row.append(f"{float(val):.1f}%")
        formatted_data.append(formatted_row)
    return np.array(formatted_data)

# Plot heatmap with custom percentage annotations
plt.figure(figsize=(10, 6))  # Larger plot size to accommodate more boxes
sns.heatmap(pivot_table_percentage, annot=annotate_percentage(pivot_table_percentage), fmt="", cmap='coolwarm', cbar=True)  # fmt="" to avoid seaborn's internal formatting
plt.title('Heatmap: Variable 1 (Buckets) vs Variable 2 (Buckets) with Percentages')
plt.xlabel('Variable 2 (Buckets)')
plt.ylabel('Variable 1 (Buckets)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a sample dataset
np.random.seed(0)
data_size = 10
variable1 = np.random.rand(data_size)
variable2 = np.random.rand(data_size)

# Create a DataFrame
df = pd.DataFrame({
    'Variable 1': variable1,
    'Variable 2': variable2
})

# Compute correlation matrix for heatmap
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Two Variables')
plt.show()
```
